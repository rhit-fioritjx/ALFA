```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
maml_df = pd.read_csv('../logs/summary_statistics.csv')
maml_df.head()
```

```python
alfa_maml_df = pd.read_csv('../../alfa+maml_medium/logs/summary_statistics.csv')
alfa_maml_df.head()
```

```python
fig, ax = plt.subplots()
ax.plot(maml_df.epoch,maml_df.val_loss_mean,label='maml validation loss')
ax.fill_between(maml_df.epoch,maml_df.val_loss_mean-maml_df.val_loss_std,maml_df.val_loss_mean+maml_df.val_loss_std,alpha=.7)
ax.plot(alfa_maml_df.epoch,alfa_maml_df.val_loss_mean,label='alfa+maml validation loss')
ax.fill_between(alfa_maml_df.epoch,alfa_maml_df.val_loss_mean-alfa_maml_df.val_loss_std,alfa_maml_df.val_loss_mean+alfa_maml_df.val_loss_std,alpha=.7)
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(maml_df.epoch,maml_df.val_loss_mean,label='maml validation loss')
# ax.fill_between(maml_df.epoch,maml_df.val_loss_mean-maml_df.val_loss_std,maml_df.val_loss_mean+maml_df.val_loss_std,alpha=.7)
ax.plot(alfa_maml_df.epoch,alfa_maml_df.val_loss_mean,label='alfa+maml validation loss')
# ax.fill_between(alfa_maml_df.epoch,alfa_maml_df.val_loss_mean-alfa_maml_df.val_loss_std,alfa_maml_df.val_loss_mean+alfa_maml_df.val_loss_std,alpha=.7)
ax.legend()
```

```python
maml_df['cum_run_time'] = [maml_df.epoch_run_time[:i].sum() for i in range(100)]
alfa_maml_df['cum_run_time'] = [alfa_maml_df.epoch_run_time[:i].sum() for i in range(100)]
```

```python
fig, ax = plt.subplots()
ax.plot(maml_df.cum_run_time,maml_df.val_loss_mean,label='maml validation loss')
ax.fill_between(maml_df.cum_run_time,maml_df.val_loss_mean-maml_df.val_loss_std,maml_df.val_loss_mean+maml_df.val_loss_std,alpha=.7)
ax.plot(alfa_maml_df.cum_run_time,alfa_maml_df.val_loss_mean,label='alfa+maml validation loss')
ax.fill_between(alfa_maml_df.cum_run_time,alfa_maml_df.val_loss_mean-alfa_maml_df.val_loss_std,alfa_maml_df.val_loss_mean+alfa_maml_df.val_loss_std,alpha=.7)
ax.legend()
```
