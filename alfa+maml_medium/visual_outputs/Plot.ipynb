```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../logs/summary_statistics.csv')
df.head()
```

```python
fig, ax = plt.subplots()
ax.plot(df.epoch,df.train_loss_mean,label='train loss')
ax.fill_between(df.epoch,df.train_loss_mean-df.train_loss_std,df.train_loss_mean+df.train_loss_std,alpha=.7)
ax.plot(df.epoch,df.val_loss_mean,label='validation loss')
ax.fill_between(df.epoch,df.val_loss_mean-df.val_loss_std,df.val_loss_mean+df.val_loss_std,alpha=.7)
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(df.epoch,df.train_accuracy_mean,label='train accuracy')
ax.fill_between(df.epoch,df.train_accuracy_mean-df.train_accuracy_std,df.train_accuracy_mean+df.train_accuracy_std,alpha=.7)
ax.plot(df.epoch,df.val_accuracy_mean,label='validation accuracy')
ax.fill_between(df.epoch,df.val_accuracy_mean-df.val_accuracy_std,df.val_accuracy_mean+df.val_accuracy_std,alpha=.7)
ax.legend()
```
